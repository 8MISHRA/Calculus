It is related to calculus in arc length.

```python
# Importing some required libraries and modules

from scipy.integrate import quad
import math as m
import sympy as sp
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
# defining function to integrate
def f(x):
  return ((((a**2)*(m.sin(x))**2) + (b**2)*(m.cos(x))**2))**0.5
```

```python
a_list = []     # list of major axis
b_list = []     # list of minor axis
f_list = []     # list of function value at given major and minor axes

a = 2
while a <= 4:
  b = 1
  while b < 2:
    b += 0.1
    result, error = quad(f, 0, 2*m.pi)
    f_list.append(result)
    a_list.append(round(a,2))
    b_list.append(round(b,2))
  a += 0.1
```

```python
# Drawing the figure
fig = plt.figure()
ax = plt.axes(projection = '3d')

# Data to be plotted
ax.plot3D(a_list, b_list, f_list, 'blue')

# Some titles and lables
ax.set_xlabel('b-value')
ax.set_ylabel('a-value')
ax.set_zlabel('Function Value ')
ax.set_title('Plotting for ellipse integral')
```

```python
# Accumulators

a_list = []     # list of major axis
b_list = []     # list of minor axis
f_list = []     # list of function value at given major and minor axes
list_1 = []     # list of Calculated by formula 1
list_2 = []     # list of Calculated by formula 2
list_3 = []     # list of Calculated by formula 3
list_4 = []     # list of Calculated by Ramanujan formula

x_axis = []     # list of x axis as a/b
a = 2
while a <= 4:
  b = 1

  result, error = quad(f, 0, 2*m.pi)
  f_list.append(result)

  value_1 = 2*m.pi* ((a+b)/2)                            # formula 1
  list_1.append(value_1)
  value_2 = 2*m.pi* (((a*a+b*b)/2)**0.5)                 # formula 2
  list_2.append(value_2)
  value_3 = m.pi* ((6*a)/5+(3*b)/4)                      # formula 3
  list_3.append(value_3)
  value_4 = m.pi*(3*(a+b)-((3*a + b)*(a + 3*b))**0.5)    # formula 4(Ramanujan formula)
  list_4.append(value_4)

  a_list.append(round(a,2))

  b_list.append(round(b,2))
  
  x_axis.append(a/b)
  a += 0.1

```

```python
# plotting some graph

plt.plot(x_axis, list_1, label = 'Graph for 2pi*a+b/2')
plt.plot(x_axis, list_2, label = '2*m.pi* (((a*a+b*b)/2)**0.5)')
plt.plot(x_axis, list_3, label = 'm.pi* ((6*a)/5+(3*b)/4)')
plt.plot(x_axis, list_4, label = 'Ramanujan')

plt.legend(title = 'Comparison all the formulae with the python-calculated formula')
plt.show()
```

```python
# Calculating error percentage

i = 0 
dif1 = []
dif2 = []
dif3 = []
dif4 = []
while i<len(a_list):
  dif_1 = abs(list_1[i] - f_list[i])/(f_list[i])*100
  dif1.append(dif_1)
  dif_2 = abs(list_2[i] - f_list[i])/(f_list[i])*100
  dif2.append(dif_2)
  dif_3 = abs(list_3[i] - f_list[i])/(f_list[i])*100
  dif3.append(dif_3)
  dif_4 = abs(list_4[i] - f_list[i])/(f_list[i])*100
  dif4.append(dif_4)
  i += 1
print(len(dif1), len(dif2), len(dif3), len(dif4))
```

```python
# Plotting percentage error
plt.plot(x_axis, dif1, label = 'Difference of error % of 2pi*a+b/2 with f')
plt.plot(x_axis, dif2, label = 'Difference of error % of 2*m.pi* (((a*a+b*b)/2)**0.5) with f')
plt.plot(x_axis, dif3, label = 'Difference of error % of m.pi* ((6*a)/5+(3*b)/4) with f')
plt.plot(x_axis, dif4, label = 'Difference of error % of Ramanujan with f')

plt.legend(title = '%age error graph w.r.t different functions')
plt.show()
```
