Error measure for analytical and mathematical data of derivative of exponential function

```python
# Importing some required modules

import math 
import matplotlib.pyplot as plt
```

```python
t_list = []
x_list = []

x_hat = 1
t = 0
del_t = 0.1

while t < 5:
  t_list.append(t)

  analytical = math.exp(t)
  
  x_hat = x_hat + x_hat*del_t
  # Finding the error b/w the analytical and symbolic derivative of exp
  difference = (analytical - x_hat)/analytical

  x_list.append(difference)
  
  t += 0.1
```

```python
plt.plot(t_list, x_list)
```
