Differentiation:
1) Numerical; it gives approximate answer
2) Symbolic; it gives exact answer
3) Automatic differentiation; it also gives exact answer


```python
# importing some required modules
import autograd.numpy as np
from autograd import grad
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import plotly.express as px
```

Derivative in one variable

```python
# Function to be differentiated

def my_function(x):
  return np.sin(x)

# Find the gradient of my_function
gradient_func = grad(np.sin)
gradient = gradient_func(2.0)

print(gradient)
```

Finding the derivative of $cosx$ using all the three methodes mentioned above

1)Using symbolic :-

```python
i = 0
cos_list = []
x1_list = []
while i <= 8 * np.pi:
  x1_list.append(i)
  j = np.cos(i)
  cos_list.append(j)

  i += 0.2
```

2) Using numerical one
### $$f\prime(x) = \frac {sin(x + Δx) - sin(x)} {Δx}$$

```python
delta_x = 0.1
x2_list = []
num_list = []

x = 0
while x <= 8 * np.pi:
  x2_list.append(x)
  j = (np.sin(x+delta_x) - np.sin(x))/delta_x
  num_list.append(j)

  x += 0.2
```

3)Using Automatic :-

```python
i = 0
x3_list = []
auto_list = []

while i <= 8*np.pi:
  x3_list.append(i)
  gradient_func = grad(my_function)
  gradient = gradient_func(float(i))
  auto_list.append(gradient)

  i += 0.5
```

```python
# Plotting comparison for the three methods

plt.plot(x1_list, cos_list, label = 'symbolic')
plt.plot(x2_list, num_list, label = 'numerical')
plt.plot(x3_list, auto_list, label = 'autograd')

plt.legend()
plt.show()
```

```python
gradient_func = grad(np.sin)

eta = 0.1
xn = 1.0
gradient = gradient_func(xn)


while abs(gradient) >= 1e-4:
  xn = xn - eta*gradient_func(xn)
  gradient = gradient_func(xn)

print(np.sin(xn))
```

Gradient descent for the function: $$f(x)=x^2+y^2$$

```python
def myfunc(X):
  x,y = X
  return x**2 + y**2

gradient_func = grad(myfunc)
```
