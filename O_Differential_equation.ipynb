```python
import math
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

let's talk about ODE (ordinery defferantial equation)
ODE:- It is the simple derivative of a function which provides us its slope 

```python
t_list = []
x_list = []
t = 0 
while t < 5:
  t_list.append(t)
  x_list.append(math.exp(t))
  t += 0.01
  
```

```python
# fig = px.line(x = t_list, y = x_list)
# fig.show()
plt.plot(t_list, x_list)
```

```python
# Eulers' Method

tlist = []
ylist = []
x_hat = 1
t = 0
while t < 5:
  tlist.append(t)
  x_hat = x_hat + x_hat*(0.01)
  ylist.append(x_hat)

  t += 0.01
```

```python
plt.plot(tlist,ylist)
```

```python
# Merging the above two graphs

plt.plot(t_list, x_list)
plt.plot(tlist,ylist)
```

```python
# Plotting the graph for error %age

err_list = []
for i in range(len(x_list)):
  err = abs((x_list[i] - tlist[i])/x_list[i])
  err_list.append(err)

plt.plot(x_list, err_list)
```
